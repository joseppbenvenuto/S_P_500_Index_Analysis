# S&P 500 Index Investment Analysis

<h3>Objective</h3>

During this analysis, The S&P 500 Index closing prices will be explored within four decades. The objective is to analyze the index and determine if the index is an invaluable long term investment.

Decade 1: 1979 - 1989<br>
Decade 2: 1989 - 1999<br>
Decade 3: 1999 - 2009<br>
Decade 4: 2009 - 2019<br>

The analysis will seek to understand the compounded returns (not including dividends) the index has produced in the above time frames. 

Within each time frame, t-tests will be performed to see if the prices are statistically significant. The goal is to observe if the closing prices are higher or lower between the time frames to determine the risk of the investment. 

Further, Monte Carlo simulations will be performed to determine risk. Using the Monte Carlo results, an average compounded return will be calculated to observe the simulated annual returns in addition to the actual compounded results in each time frame.

<br>

```python
# Import relevant packages
import pandas as pd 
# Import relevant defined funtions
from ipynb.fs.full.Stats_Monte_Carlo_Functions import *

# Import data sets
clean_com = pd.read_csv('Data/Comprehensive_Data.csv')
clean_com = clean_com.drop(['Unnamed: 0'], axis = 1)
da_1 = pd.read_csv('Data/Decade1.csv')
da_1 = da_1.drop(['Unnamed: 0'], axis = 1)
da_2 = pd.read_csv('Data/Decade2.csv')
da_2 = da_2.drop(['Unnamed: 0'], axis = 1)
da_3 = pd.read_csv('Data/Decade3.csv')
da_3 = da_3.drop(['Unnamed: 0'], axis = 1)
da_4 = pd.read_csv('Data/Decade4.csv')
da_4 = da_4.drop(['Unnamed: 0'], axis = 1)
```

Import the necessary data for the analysis.

Earlier recessions took place in 1980 - 1983 due to the energy crisis (high oil prices) with high inflation and interest rates. This is not depicted as aggressively in the pricing of the S&P 500 index prices as the dot-com bubble and the financial crisis. 

The S&P 500 Index has shown to grow and provide solid returns, via price increases, even having gone through significant adverse economic events. Being invested in the S&P 500 Index over the long term proves to be a solid investment.

The Annual Real U.S GDP has shown to grow steadily over the four decades measured and has grown at a compound rate of 3% annually. There are flat lines and dips in the early 1980s and late in the first decade of the 21 century (2000s), this can be due to the recessions explained above. 

Additionally, we can observe the pricing of the stock market over time (S&P 500 Index) as adverse economic events took place. while the effects of the early 1980s may seem small in the line plots above, this period was known to be the worst recession since the great depression of 1929-1939 during its time. During this period, however, prices seemed to remain steady. 

Looking forward to the dot-com bubble and the financial crisis, stock prices became very volatile.

If the investor invested in the S&P 500 index and remained unphased by the volatility or adverse economic events (invested capital was not needed to finance everyday life and little to no debt), the investor would have capitalized on the resiliency and strength of the American economy. <h3>S&P 500 Index Closing Prices Insights for Decade 1 and Decade 2</h3>

```python
# T-tests between Decade 1 and Decade 2
series_stats_gen2('Decade 1',da_1['s&p_closing_prices'],'Decade 2',da_2['s&p_closing_prices'], 
                  'Decade 1 Prices vs Decade 2 Prices')
print('\n')
# Monte Carlo simulation of S&P 500 Index prices between Decade 1 and Decade 2
price_sims(da_1,'s&p_closing_prices', 'Decade 1',da_2,'s&p_closing_prices','Decade 2',1000,20)
```

The null hypothesis stating there is no difference between the means of the S&P 500 Index closing prices for Decade 1 and Decade 2 is rejected. 

The prices between Decade 1 and Decade 2 are statistically significant at p < 0.05. With 95% confidence, the difference can be found between -386.29 & -366.81. This shows that prices were lower in Decade 1 in comparison to Decade 2 and therefore, on average, the investor would have turned a profit if the investor was invested in the S&P 500 Index during Decade 1 through Decade 2.

To further illustrate the point above, a Monte Carlo simulation was performed with sample sizes of 1,000 individual random prices from both decades'. Using the random prices, an average compounded return was calculated and the following results were determined; if the investor invested in Decade 1 through Decade 2, the investor would have earned an average compounded return of 6%. We also saw that the prices in Decade 1, at a rate of 100%, were lesser than Decade 2 prices.

Additionally, if The investor bought the S&P 500 Index at the beginning of Decade 1 and sold at the end of Decade 2 (20 years), the investor would have earned an actual compounded return of 14%.

If the investor invested in Decade 1 through Decade 2, the investor would have earned satisfactory returns with a high margin of safety.
