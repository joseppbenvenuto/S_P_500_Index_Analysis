# S&P 500 Index Investment Analysis

<h3>Description</h3>

In the below analysis, the S&P 500 Index's closing prices will be explored within a four-decade time frame. The objective is to analyze the index and determine if the index is an invaluable long term investment.

Four decades:

Decade 1: 1979 - 1989<br>
Decade 2: 1989 - 1999<br>
Decade 3: 1999 - 2009<br>
Decade 4: 2009 - 2019<br>

The analysis will seek to understand the compounded returns (not including dividends) the index has produced in the above within time frames. 

A/B will be used to infer if the difference in means of prices between decades is statistically significant. The goal is to observe if the closing prices are higher or lower between the time frames and determine the investment risk. 

If the prices are higher in each new decade and statistically significant, intuitively, this will be determined to be a positive sign and that the index is producing profits; the opposite is true for the opposite results.

In addition to the above, the analysis will observe trends in economic data regarding U.S Real GDP, S&P 500 Index EPS, Ten Year U.S Treasury Bond Yields, and Average Annual U.S Inflation to aid in the analysis.

The analysis can be taken advantage of by any individual or entity looking for a valuable investment opportunity.

<h3>Trend Insights</h3>

<h4>S&P 500 Index Daily Closing Prices vs Years</h4><br>
The S&P 500 Index closing prices have increased within the four decades measured (1979-2019) and has grown annually at a compound rate of 9%. The slope is not a consistent rise due to volatile periods in the mid-1990s through 2010 and can be explained by the existence of the dot-com bubble (1995-2001) and the financial crisis (2007-2009).

Earlier recessions took place in 1980 - 1983 due to the energy crisis (high oil prices) with high inflation and interest rates. This is not depicted as aggressively in the pricing of the S&P 500 index prices as the dot-com bubble and the financial crisis. 

The S&P 500 Index has shown to grow and provide solid returns, via price increases, even having gone through significant adverse economic events. Being invested in the S&P 500 Index over the long term appears to be a profitable investment.  

<h4>U.S Real GDP vs Years</h4><br>
The Annual Real U.S GDP has shown to grow steadily over the four decades measured and has grown at a compound rate of 3% annually. There are flat lines and dips in the early 1980s and late in the first decade of the 21 century (2000s), this can be due to the recessions explained above. 

Additionally, we can observe the pricing of the stock market over time (S&P 500 Index) as adverse economic events took place. while the effects of the early 1980s may seem small in the line plots above, this period was known to be the worst recession since the great depression (1929-1939) during its time. During this period, however, prices seemed to remain steady. 

Looking forward to the dot-com bubble and the financial crisis, stock prices became very volatile.

If the investor invested in the S&P 500 index and remained unphased by the volatility or adverse economic events (invested capital was not needed to finance everyday life and little to no debt), the investor would have capitalized on the resiliency and strength of the American economy. The investor would have profited from the investment with a high margin of safety.
<br>

<h4>S&P 500 Index EPS vs Years</h4><br>
The S&P 500 Index EPS has to grown within the four decades measured. There is more volatility in comparison to annual real U.S GDP and the S&P 500 Index's prices. 

The earnings on a per-share basis have been quite volatile, however, the annual compounded rate of growth for the EPS is 3% annually. This continues to build the case that the longer time horizon, through volatile events, increases profits and that the index, which represents the strength of American business, is a profitable investment to be in overtime.
<br>

<h4>Ten Year U.S Treasury Bond Yields vs Years</h4><br>
The Ten Year U.S Treasury Bond Yields have shown to decline steadily over the four decades. The annual compound rate of growth for the bond yields is -4% annually. 

Due to the known adverse relationship asset prices have with interest rates, the steady drop in rates may be the reason for the S&P 500 Index continued price increases

On the contrary, the index, since its inception in 1959, has returned 7% compounded annually. During the time frame of the index's existence, interest rates have fluctuated up and down and have been through several recessions. Taking the latter into account, interest rates cannot be the main reason why the S&P 500 Index prices have increased but may be the reason in the shorter term. 
<br>

<h4>Average Annual U.S Inflation vs Years</h4><br>
The Average Annual U.S Inflation has been shown to decline steadily over the four decades measured and has grown annually at a compound rate of growth of -4%. Like interest rates, bond yields have a known adverse affect on stock prices and my be, the reason in the shorter term, why the S&p 500 Index has continued its steady price increase.

<h3>S&P 500 Index Closing Prices Insights for Decade 1 and Decade 2</h3>

```python
# T-tests between Decade 1 and Decade 2
series_stats_gen2(string = 'Decade 1', column = da_1['s&p_closing_prices'], string2 = 'Decade 2',
                  column2 = da_2['s&p_closing_prices'], t = 'Decade 1 Prices vs Decade 2 Prices', years = 20,
                  length = 3, width = 10)
```

The null hypothesis stating that there is no difference between the means of the S&P 500 Index closing prices for Decade 1 and Decade 2 is rejected. 

The prices between Decade 1 and Decade 2 are statistically significant at p < 0.05. With 95% confidence, the difference can be found between -386.29 & -366.81 (distribution: 166.9, 176.64, 186.38). This shows that prices were lower in Decade 1 in comparison to Decade 2 and therefore, on average, the investor would have turned a profit if the investor was invested in the S&P 500 Index during Decade 1 through Decade 2.

Additionally, if The investor bought the S&P 500 Index at the beginning of Decade 1 and sold at the end of Decade 2 (20 years), the investor would have earned an actual compounded return of 14%.

If the investor invested in Decade 1 through Decade 2, the investor would have earned satisfactory returns with a high margin of safety.
